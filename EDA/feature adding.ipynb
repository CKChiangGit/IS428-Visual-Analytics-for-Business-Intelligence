```python
import pandas as pd
import matplotlib.pyplot as plt
import openpyxl as px
```

```python
# Datasets
path = "./Datasets/"

# main
main_df = pd.read_excel('CO2 Emissions (Cleaned).xlsx');

# population
pop_df = pd.read_excel(path + 'Population Total.xlsx');

# GDP
gdp_df = pd.read_excel(path + 'GDP (current US$).xlsx');

# Imports and exports
imp_df = pd.read_excel(path + 'Imports of goods and services (% of GDP).xlsx');
exp_df = pd.read_excel(path + 'Exports of goods and services (current US$).xlsx');
```

```python
# Drop 'Country Code', 'Indicator Name', and 'Indicator Code' columns
pop_df = pop_df.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code'])

# Melt the DataFrame, using 'Country Name' as id_vars and the year columns for melting
pop_df_melt = pop_df.melt(id_vars=['Country Name'], 
                          var_name='Year', 
                          value_name='Population')

# Convert the 'Year' column to integer
pop_df_melt['Year'] = pop_df_melt['Year'].astype(int)

from difflib import get_close_matches

# Helper function to find the best match
def find_best_match(country):
    matches = get_close_matches(country, pop_df_melt['Country Name'].unique(), n=1, cutoff=0.1)
    return matches[0] if matches else country

# Apply the function to 'Country' column
main_df['Country Match'] = main_df['Country'].apply(find_best_match)

# # Step 3: Merge main_df and the transformed pop_df
# Merge main_df and pop_df_melt using 'Country Match' and 'Year' in main_df
merged_df = pd.merge(main_df, pop_df_melt, 
                     left_on=['Country Match', 'Year'], 
                     right_on=['Country Name', 'Year'], 
                     how='left')

# Drop the 'Country Match' and 'Country Name' columns
merged_df = merged_df.drop(columns=['Country Match', 'Country Name'])

# Replace the 'Population' column in main_df
merged_df['Population'] = merged_df['Population_y'].where(merged_df['Population_y'].notna(), merged_df['Population_x'])

# Identify the countries in main_df that do not have matching records in pop_df
no_match_countries = merged_df[merged_df['Population_y'].isna()]['Country'].unique().tolist()
print(no_match_countries)

# Drop the 'Population_x' and 'Population_y' columns if no_match_countries is empty
if not no_match_countries:
    merged_df = merged_df.drop(columns=['Population_x', 'Population_y'])
```
